```python
import matplotlib.pyplot as plt
ds_sorted = ds.sort_values(by="likes", ascending=False)
most_liked_videos = ds_sorted.iloc[:10]
likes = most_liked_videos["likes"]
titles = most_liked_videos["title"]
explode = [0.1 if i == most_liked_videos["likes"].idxmax() else 0 for i in range(len(likes))]
plt.figure(figsize=(8, 8))
plt.pie(likes, labels=titles, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal') 
plt.title("Distribution of Likes for Top 10 Most Liked Videos")
plt.show()
```

```python
category_stats = ds.groupby('title')[['likes', 'dislikes', 'view_count']].sum()
most_popular_category = category_stats.sum(axis=1).idxmax()
most_popular_stats = category_stats.loc[most_popular_category]
print("Most Popular Video:", most_popular_category)
print("Total Likes:", most_popular_stats['likes'])
print("Total Dislikes:", most_popular_stats['dislikes'])
print("Total Views:", most_popular_stats['view_count'])
```

```python
ds['publishedAt'] = pd.to_datetime(ds['publishedAt'])
ds['trending_date'] = pd.to_datetime(ds['trending_date'])
ds['publishedAt'] = pd.to_datetime(ds['publishedAt'])
ds['trending_date'] = pd.to_datetime(ds['trending_date'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
ds['publishedAt'] = pd.to_datetime(ds['publishedAt'])
ds.set_index('publishedAt', inplace=True)
result = seasonal_decompose(ds['likes'], model='additive', period=30)  
plt.figure(figsize=(12, 6))
result.plot()
plt.show()
```

```python
most_viewed_channel = ds.groupby('channelTitle')['view_count'].sum().idxmax()
most_viewed_channel_view_count = ds.groupby('channelTitle')['view_count'].sum().max()
print(f"The most viewed channel is {most_viewed_channel} with a total view count of {most_viewed_channel_view_count}.")
```

```python
channel_views = ds.groupby('channelTitle')['view_count'].sum()
channel_views = channel_views.sort_values(ascending=False)
N = 20  
top_channels = channel_views[:N]
plt.figure(figsize=(12, 6))
top_channels.plot(kind='barh', color='skyblue')
plt.xlabel('Total View Count')
plt.ylabel('Channel Title')
plt.title(f'Top {N} Most Viewed Channels')
plt.gca().invert_yaxis()  
plt.show()
```

```python
ds['trending_date'] = pd.to_datetime(ds['trending_date'], format='%y.%d.%m')
ds['trending_year'] = ds['trending_date'].dt.year
channel_views_by_year = ds.groupby(['trending_year', 'channelTitle'])['view_count'].sum().reset_index()
most_viewed_channel_by_year = channel_views_by_year.loc[channel_views_by_year.groupby('trending_year')['view_count'].idxmax()]
plt.figure(figsize=(12, 6))
for year, group in most_viewed_channel_by_year.groupby('trending_year'):
    plt.bar(str(year), group['view_count'], label=group['channelTitle'].values[0])

plt.xlabel('Year')
plt.ylabel('Total View Count')
plt.title('Most Viewed Channel by Year')
plt.legend(loc='best')
plt.show()
```

```python
dislike_video = ds.groupby('title')["dislikes"].sum()
dislike_video = dislike_video.sort_values(ascending=False)
N = 10  
top_channels = dislike_video[:N]
plt.figure(figsize=(12, 6))
top_channels.plot(kind='barh', color='skyblue')
plt.xlabel('Most dislike video ')
plt.ylabel('video Title')
plt.title(f'Top {N} Most dislike video')
plt.gca().invert_yaxis()  
plt.show()
```
